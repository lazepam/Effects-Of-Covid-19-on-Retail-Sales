## 2. Extract-Transform-Load

    The ETS process understands the relationship between datasets from different systems and is the process by which we move data to various storage types for analysis.
    
    It represents capturing data in a way that computer programs cannot understand and converting it into a format that programs can understand.
    
    The letters in the ETL abbreviation represent the abbreviation of the applied process as E mean Extract. At this stage, raw data is obtained from the source or sources.
    
    T mean Transform, at this stage, raw data obtained from different sources is purified from unnecessary data or data is organized according to the formulation or project to be applied. So the raw data becomes processed for use in programming.
    
    L mean Load. In this step, the collected and processed data is stored for use in the program to be used.

```python
import pandas as pd
```

```python
df = pd.DataFrame({'date':liste,'men_sale':men_sale, 'women_sale':women_sale})
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d') 
```

```python
df.groupby(df['date'].dt.to_period('Q'))['men_sale'].sum() 
```

```python
df.info()
```

```python
df.groupby(df['date'].dt.to_period('Q'))['men_sale'].sum().plot()
plt.xlabel('date', fontsize=10)
plt.ylabel('men_sales', fontsize=10)
plt.show()
```

### 3.3 Explore Trends

    In an uptrend market, it is of great importance to determine the lower limit of the price movement, as the transactions take place in the buying direction. When prices break the trend line, the trend starts to weaken and the direction of the trend changes. The trend line that limits price changes from below is called the “support line”.

    Looking at the uptrend price chart, it is seen that the bottom points of the fluctuations are lined up in the upward direction. This line, which limits the trend from the downside and passes through the minimum points of price fluctuations, is called the trend line.

    A horizontal trend is a trend pattern where there is no specific trend direction and price changes occur at a horizontal level.

    In our analysis, while there was a slight horizontal increase during the period we examined, a rapid downward trend was observed in the last period.

```python
df.groupby(df['date'].dt.to_period('Q'))['women_sale'].sum().plot()
plt.xlabel('date', fontsize=10)
plt.ylabel('women_sale', fontsize=10)
plt.show()
```

```python
#create scatterplot
plt.scatter(df['date'], df['women_sale'])
plt.show()
```
